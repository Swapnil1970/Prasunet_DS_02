```python
#importing packages
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#loading the training dataset into dataframe df
df = pd.read_csv("D:\\Users\\shind\\Downloads\\train.csv")
df
```

```python
#retrieving number of columns in dataset vertically
df.columns.to_frame().reset_index().drop(0, axis=1)
```

```python
#renaming some columns with relevent names
df.rename(columns={
    'PassengerId' : 'Passenger_ID',
    'Pclass' : 'Passenger_Class',
    'Sex' : 'Gender',
    'SibSp' : 'Siblings_Spouses_Aboard',
    'Parch' : 'Parents_Children_Aboard',
    'Ticket' : 'Ticket_number',
}, inplace=True)
df
```

```python
#checking for null values
df.isna().sum()
```

```python
#deleting Cabin column since it contains a lot of null values
df.drop(columns='Cabin',inplace=True)
```

```python
#treating other columns which also contain missing values
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Age'] = df['Age'].fillna(df['Age'].median())
df.drop(columns=['Name','Ticket_number'],inplace=True)
```
